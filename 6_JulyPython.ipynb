```python
years=[1950,1960,1970,1980,1990,2000,2021,2020]
company_x =[2.5,3.5,2.8,3.1,4.0,2.9,4.5,4.8]

years=[1950,1960,1970,1980,1990,2000,2021,2020]
company_y=[2.2,3.5,2.5,3.5,4.5,2.1,4.4,4.5]

font1={'family':'serif','color':'blue','size':12}
font2={'family':'serif','color':'blue','size':14}
       
plt.subplot(1, 2, 1)
plt.plot(years,company_x,color='blue' ,marker='o', ms='10', mfc= 'green', mec='black')
plt.grid(linestyle = 'dashed' , color='black')
plt.xlabel("years")
plt.ylabel("Company_x_Captial")
plt.title("Company_X_campital Over Years" ,fontdict= font1)

plt.subplot(1,2,2)
plt.plot(years,company_y, color='red' ,marker='D', ms='10', mfc= 'green', mec='red')
plt.grid(linestyle = 'solid' , color='black')
plt.xlabel("years")
plt.ylabel("Company_y_Captial")
plt.title("Company_Y_campital Over Years" ,fontdict=font2)

plt.subplots_adjust(wspace =1)
plt.suptitle("Capital Comparison of companies Over years")
plt.show()
```

# Homework- search the popluation pyramid of 10 countries - Russia, america,china,canada, # indonesia,india,pakistan,brazil,england,bangladesh, show using line plot,scatter plot,bargraph,horizontal baragraph using aall the attributes of functions

```python

data = {
    'Country': ['Russia', 'USA', 'China', 'Canada', 'Indonesia', 'India', 'Pakistan', 'Brazil', 'UK', 'Bangladesh'],
    '0-14': [17.1, 18.3, 17.8, 16.0, 25.3, 27.3, 35.4, 21.0, 17.6, 27.9],
    '15-24': [9.6, 13.0, 12.5, 12.0, 16.9, 18.1, 20.1, 16.0, 11.8, 19.4],
    '25-54': [44.3, 39.8, 48.4, 41.0, 42.4, 41.0, 36.7, 43.0, 39.7, 39.1],
    '55-64': [13.4, 12.9, 10.0, 14.0, 7.9, 8.4, 5.9, 10.0, 12.3, 7.3],
    '65+': [15.6, 16.0, 11.3, 17.0, 7.5, 5.2, 1.9, 10.0, 18.6, 6.3]
}

df = pd.DataFrame(data)

# Line Plot
df.plot(x='Country', y=['0-14', '15-24', '25-54', '55-64', '65+'], marker='o')
plt.title('Population Distribution by Age Group - Line Plot')
plt.xlabel('Country')
plt.ylabel('Percentage')
plt.show()

# Scatter Plot
df.plot(x='Country', y=['0-14', '15-24', '25-54', '55-64', '65+'], kind='scatter', c='b')
plt.title('Population Distribution by Age Group - Scatter Plot')
plt.xlabel('Country')
plt.ylabel('Percentage')
plt.show()

# Bar Plot
df.plot(x='Country', y=['0-14', '15-24', '25-54', '55-64', '65+'], kind='bar')
plt.title('Population Distribution by Age Group - Bar Graph')
plt.xlabel('Country')
plt.ylabel('Percentage')
plt.show()

# Horizontal Bar Plot
df.plot(x='Country', y=['0-14', '15-24', '25-54', '55-64', '65+'], kind='barh')
plt.title('Population Distribution by Age Group - Horizontal Bar Graph')
plt.xlabel('Percentage')
plt.ylabel('Country')
plt.show()
```
