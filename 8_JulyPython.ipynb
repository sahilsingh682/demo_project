```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
file_path = 'C:\\Users\\kanik\\Downloads\\delhi.csv'
delhi = pd.read_csv(file_path)
delhi.head()
```

```python
file_path = 'C:\\Users\\kanik\\Downloads\\mumbai.csv'
mumbai = pd.read_csv(file_path)
mumbai.head()
```

```python
fig , ax=plt.subplots(3,2,figsize=(13,15))
font1={'family':'serif','color':'darkred','size':10}
font2={'family':'serif','color':'blue','size':12}

ax[0,0].plot(delhi['Month'],delhi['Minimum'],color='r',marker='o')
ax[0, 0].tick_params(axis='x', rotation=90)
ax[0,0].set_xlabel("Month",fontdict=font1)
ax[0,0].set_ylabel("Temp in C",fontdict=font2)

ax[0,1].plot(mumbai['Month'],mumbai['Minimum'],color='r',marker='D')
ax[0,1].tick_params(axis='x', rotation=90)
ax[0,1].set_xlabel("Month",fontdict=font1)
ax[0,1].set_ylabel("Temp in C",fontdict=font2)

ax[1,0].plot(delhi['Month'],delhi['Maximum'],color='b',marker='o')
ax[1, 0].tick_params(axis='x', rotation=90)
ax[1,0].set_xlabel("Month",fontdict=font1)
ax[1,0].set_ylabel("Temp in C",fontdict=font2)

ax[1,1].plot(mumbai['Month'],mumbai['Maximum'],color='b',marker='D')
ax[1, 1].tick_params(axis='x', rotation=90)
ax[1,1].set_xlabel("Month",fontdict=font1)
ax[1,1].set_ylabel("Temp in C",fontdict=font2)

ax[2,0].plot(delhi['Month'],delhi['Average'],color='g',marker='o')
ax[2, 0].tick_params(axis='x', rotation=90)
ax[2,0].set_xlabel("Month",fontdict=font1)
ax[2,0].set_ylabel("Temp in C",fontdict=font2)

ax[2,1].plot(mumbai['Month'],mumbai['Average'],color='g',marker='D')
ax[2,1].tick_params(axis='x', rotation=90)
ax[2,1].set_xlabel("Month",fontdict=font1)
ax[2,1].set_ylabel("Temp in C",fontdict=font2)

plt.suptitle("Temp Comparision of Delhi and Mumbai")
# plt.xticks(rotation=90)
# plt.tight_layout()
plt.show()
```
