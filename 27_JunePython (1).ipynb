# DataFrame

# dataframe is a datastructure that stores the data in two dimensional form.

```python
# mark = np.arange(61,91).reshape(6,5)
mark = np.arange(61,91)
shuffle(mark)
mark= mark.reshape(6,5)
print(mark)
```

```python
print(mark)
```

```python
names = ['nitin','manvi','kanika','sahil','uttam','prince']
subject=['hindi','english','maths','physics','chemistry']
```

```python
grade= pd.DataFrame(data=mark , index=names , columns=subject)
```

```python
grade
```
