# joining dataframe with different columns

```python
dict3={
    'iid':[7,8,9],
    'cid':[38,40,42],
    'invoice_date':['2024-07-01','2024-07-02','2024-07-01'],
    'total':[1.98,2.56,3.6],
    'bill_city':['panipat','delhi','gurugram']
}
```

```python
inv_feb  = pd.DataFrame(dict3)
inv_feb
```

```python
pd.concat([inv_jan , inv_feb])
```

```python
pd.concat([inv_jan , inv_feb] , sort= True)
```

```python
pd.concat([inv_jan , inv_feb] , join='inner')
```
